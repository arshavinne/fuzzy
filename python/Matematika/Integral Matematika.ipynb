```python
import numpy as np
import scipy.integrate as integrate
import sympy as sp
import matplotlib.pyplot as plt

x = sp.Symbol('x')

f1 = 3*x**2 + 5*x + 2
f2 = sp.sin(x)
f3 = sp.exp(x)
f4 = x*sp.log(x)

print("Fungsi-fungsi yang akan diintegralkan:")
print("f1 =", f1)
print("f2 =", f2)
print("f3 =", f3)
print("f4 =", f4)

print("\nIntegral tak tentu dari fungsi-fungsi tersebut:")
F1 = sp.integrate(f1,x)
F2 = sp.integrate(f2,x)
F3 = sp.integrate(f3,x)
F4 = sp.integrate(f4,x)
print("F1 =", F1)
print("F2 =", F2)
print("F3 =", F3)
print("F4 =", F4)

print("\nProses perhitungan integral tak tentu dengan sympy:")
sp.pprint(sp.Integral(f1,x))
sp.pprint(F1)
sp.pprint(sp.Integral(f2,x))
sp.pprint(F2)
sp.pprint(sp.Integral(f3,x))
sp.pprint(F3)
sp.pprint(sp.Integral(f4,x))
sp.pprint(F4)

a = 0
b = 1

print("\nIntegral tentu dari fungsi-fungsi tersebut antara", a, "dan", b, ":")
I1 = sp.integrate(f1,(x,a,b))
I2 = sp.integrate(f2,(x,a,b))
I3 = sp.integrate(f3,(x,a,b))
I4 = sp.integrate(f4,(x,a,b))
print("I1 =", I1)
print("I2 =", I2)
print("I3 =", I3)
print("I4 =", I4)

print("\nProses perhitungan integral tentu dengan sympy:")
sp.pprint(sp.Integral(f1,(x,a,b)))
sp.pprint(I1)
sp.pprint(sp.Integral(f2,(x,a,b)))
sp.pprint(I2)
sp.pprint(sp.Integral(f3,(x,a,b)))
sp.pprint(I3)
sp.pprint(sp.Integral(f4,(x,a,b)))
sp.pprint(I4)

g1 = lambda x: 3*x**2 + 5*x + 2
g2 = lambda x: np.sin(x)
g3 = lambda x: np.exp(x)
g4 = lambda x: x*np.log(x)

print("\nIntegral tentu dari fungsi-fungsi tersebut antara", a, "dan", b, "dengan menggunakan scipy:")
J1, err1 = integrate.quad(g1,a,b)
J2, err2 = integrate.quad(g2,a,b)
J3, err3 = integrate.quad(g3,a,b)
J4, err4 = integrate.quad(g4,a,b)
print("J1 =", J1, "+-", err1)
print("J2 =", J2, "+-", err2)
print("J3 =", J3, "+-", err3)
print("J4 =", J4, "+-", err4)

xx = np.linspace(a,b,100)
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.plot(xx,g1(xx),label="f1")
plt.fill_between(xx,g1(xx),alpha=0.5)
plt.title("Integral tentu dari f1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.subplot(222)
plt.plot(xx,g2(xx),label="f2")
plt.fill_between(xx,g2(xx),alpha=0.5)
plt.title("Integral tentu dari f2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.subplot(223)
plt.plot(xx,g3(xx),label="f3")
plt.fill_between(xx,g3(xx),alpha=0.5)
plt.title("Integral tentu dari f3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.subplot(224)
plt.plot(xx,g4(xx),label="f4")
plt.fill_between(xx,g4(xx),alpha=0.5)
plt.title("Integral tentu dari f4")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```
