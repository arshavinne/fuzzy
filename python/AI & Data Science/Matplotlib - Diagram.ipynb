```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1,3,6,8])
ypoints = np.array([3,8,4,12])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
x = np.array([5,7,8,9,10,25,67,85,98,55])
y = np.array([6,9,12,35,50,45,64,95,84,55])

plt.scatter(x, y)
plt.show
```

```python
x = np.array([5,7,8,9,10,25,67,85,98,55])
y = np.array([6,9,12,35,50,45,64,95,84,55])
size = np.array([15,45,25,35,125,95,75,50,25,5])

plt.scatter(x, y, s=size)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 9, 7, 12])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 9, 7, 12])

plt.bar(x, y, width=0.5, color='green')
plt.show()
```

```python
x = np.array([20,30,25,25])

plt.pie(x)
plt.show()
```

```python
labels = ["Yogyakarta", "Bandung", "Jakarta", "Surabaya"]
explode = [0, 0.3, 0, 0]
x = np.array([20,30,25,25])

plt.pie(x, labels=labels, explode=explode)
plt.legend(loc="center right")
plt.show()
```
