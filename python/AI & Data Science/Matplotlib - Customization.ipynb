```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints, marker='o')
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints, ':')
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints, 'g')
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints, 'o--r')
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints, marker='o', ms='14', mec='brown', mfc='#4CAF50')
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints= np.array([3,9,27,81])

plt.plot(xpoints, ypoints)

plt.title("Data Olahraga")
plt.xlabel("Rata rata tekanan")
plt.ylabel("Pembakaran Kalori")
plt.show()
```

```python
font_title = {'family':'serif','color':'orange', 'size':20}
font_label = {'family':'serif','color':'blue', 'size':15}

xpoints = np.array([2,4,8,16])
ypoints = np.array([3,9,27,81])

plt.plot(xpoints, ypoints)

plt.title("Data Olahraga", fontdict=font_title)
plt.xlabel("Rata rata tekanan", fontdict=font_label)
plt.ylabel("Pembakaran Kalori", fontdict=font_label)
plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints = np.array([3,9,27,81])

plt.plot(xpoints, ypoints, marker='o')

plt.grid()

plt.show()
```

```python
xpoints = np.array([2,4,8,16])
ypoints = np.array([3,9,27,81])

plt.plot(xpoints, ypoints, marker='o')

plt.grid(axis='y')

plt.show()
```
