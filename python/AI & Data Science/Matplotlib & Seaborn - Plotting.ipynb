```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([0,25])
ypoints = np.array([0,175])
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([0,25])
ypoints = np.array([175,0])
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([0,25])
ypoints = np.array([0,175])

plt.plot(xpoints, ypoints, marker='*')
plt.show
```
